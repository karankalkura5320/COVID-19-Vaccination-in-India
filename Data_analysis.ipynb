```python
# Importing all the necessary library required for this project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Read the data from the datasets
data = pd.read_csv("covid-19_vaccine_data.csv")
```

### **Data Cleaning**

```python
#To fill NAN values with mean
data.fillna(data.mean(), inplace=True)
```

```python
# Taking the required value for analysis
# Total Sessions Conducted
x = []
y = 7
for i in range(139):
    x.append(y)
    y += 10
```

```python
valid_data = data.iloc[:, x]
```

```python
valid_data
```

```python
# Remove the unneccessary coloumns
valid_data.drop('Unnamed: 1347', axis=1, inplace=True)
valid_data.drop('Unnamed: 1357', axis=1, inplace=True)
valid_data.drop('Unnamed: 1367', axis=1, inplace=True)
valid_data.drop('Unnamed: 1377', axis=1, inplace=True)
valid_data.drop('Unnamed: 1387', axis=1, inplace=True)
```

```python
# Remove the first row
valid_data.drop(0, axis=0, inplace=True)
```

```python
valid_data
```

```python
valid_data = valid_data.dropna()
```

```python
X = valid_data.iloc[:, :-1]
y = valid_data.iloc[:, -1]
```

```python
valid_data
```

## **Data Splitting**

```python
from sklearn.model_selection import train_test_split as tts
```

```python
# Spliting  dataset into train and test
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=10)
```

## **Algorithm**

```python
# Import the model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```python
# Fit the model
linreg.fit(X_train, y_train)
```

# **Testing**

```python
# Predict the value
my_pred = linreg.predict(y_test)
```

```python
my_pred
```

```python
# Calculate the accuracy score
my_score = linreg.score(X_test, y_test)
```

```python
my_score
```
